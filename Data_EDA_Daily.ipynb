```python
import pandas as pd
import os
import glob
```

```python
path = '/Users/ishantchaudhary/DL_Assignment/Crypto_data_daily/'
extension = 'csv'
os.chdir(path)
result = glob.glob('*.{}'.format(extension))
```

```python
#Appending the dataset of different cryptos
dfs = list()
for f in result:
    data = pd.read_csv(f)
    # .stem is method for pathlib objects to get the filename w/o the extension

    size = len(f)
    data['Crypto'] = f[:size - 8]
    # adding percentage change in closing price 
    data['Change_close'] = data['Close'].pct_change()
    data['Change_close'] = data['Change_close'].bfill()
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)
```

```python
data = df.drop(columns=['Open Time.1','Close Time','Ignore'])
```
